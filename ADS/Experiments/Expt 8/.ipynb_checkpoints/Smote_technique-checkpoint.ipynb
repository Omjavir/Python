```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

def generate_synthetic_samples(X, y, k=5, num_synthetic=1):
    """
    Generate synthetic samples using SMOTE technique.
    
    Parameters:
    X (array-like): Feature matrix.
    y (array-like): Target labels.
    k (int): Number of nearest neighbors to consider.
    num_synthetic (int): Number of synthetic samples to generate for each minority sample.
    
    Returns:
    array-like: Feature matrix with synthetic samples.
    array-like: Target labels with synthetic samples.
    """
    synthetic_X = []
    synthetic_y = []
    
    # Separate minority and majority class samples
    minority_class = X[y == 1]
    majority_class = X[y == 0]
    
    # Fit nearest neighbors model
    nbrs = NearestNeighbors(n_neighbors=k+1).fit(minority_class)
    
    for i, sample in enumerate(minority_class):
        # Find k nearest neighbors for each minority sample
        _, indices = nbrs.kneighbors(sample.reshape(1, -1))
        
        # Generate synthetic samples
        for j in range(num_synthetic):
            nn = np.random.choice(indices[0][1:])  # Exclude the sample itself
            diff = minority_class[nn] - sample
            synthetic_sample = sample + np.random.random() * diff
            synthetic_X.append(synthetic_sample)
            synthetic_y.append(1)
    
    synthetic_X = np.array(synthetic_X)
    synthetic_y = np.array(synthetic_y)
    
    # Combine original and synthetic samples
    new_X = np.vstack((X, synthetic_X))
    new_y = np.concatenate((y, synthetic_y))
    
    return new_X, new_y

# Example data generation
np.random.seed(0)

# Class 1: Majority class
X_majority = np.random.normal(loc=[0, 0], scale=[1, 1], size=(100, 2))
y_majority = np.random.normal(loc=[0, 0], scale=[1, 1], size=(100, 2))

# Class 2: Minority class
X_minority = np.random.normal(loc=[2, 2], scale=[0.5, 0.5], size=(20, 2))
y_minority = np.random.normal(loc=[2, 2], scale=[0.5, 0.5], size=(20, 2))

# Combine data
X = np.vstack((X_majority, X_minority))
y = np.concatenate((y_majority, y_minority))

# Resample using SMOTE
X_resampled, y_resampled = generate_synthetic_samples(X, y)

print(X_resampled.shape)
print(X_majority.shape)

# Plot original and resampled data
plt.figure(figsize=(10, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Majority Class', alpha=0.6)
plt.scatter(X_resampled[y_resampled == 1][:, 0], X_resampled[y_resampled == 1][:, 1], color='green', label='Synthetic Samples', alpha=0.6)

# plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Minority Class', alpha=0.2)
plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Original and Resampled Data')
plt.show()

```
