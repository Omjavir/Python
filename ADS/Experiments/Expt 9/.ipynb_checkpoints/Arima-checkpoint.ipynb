```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv('../IT_Salary_Survey.csv')
df.head()
```

```python
def parser(x):
    return datetime.strptime(x, '%Y-%m')

def getDate(x):
    d = x.split(' ')[0: 2]
    d[1] = str(int(d[1][0])*3)
    print(parser("-".join(d)))
#     print(d)
```

```python
for series_name, series in df.items():
    if series_name == 'Country of Nationality': continue
    getDate(series_name)

new_df = pd.DataFrame()
for series_name, series in df.T.items():
#     print(series_name)
    if series_name == 1:
        new_df[series_name] = series
```
